```python
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# load dataset
dataset = pd.read_csv(r"C:\Users\HP\Desktop\Final_COP\FeatureImportance.csv", encoding='cp1252') 
dataset = dataset.drop('Unnamed: 0', axis = 1)

datasetF = dataset.drop(['PHQ'],axis=1)
```

```python
X_train = dataset.drop(['PHQ'],axis=1).values #features = X arrays 
y_train = dataset['PHQ'].values #labels = y arrays
```

```python
# define the model
model = RandomForestClassifier(class_weight='balanced_subsample',max_features=  0.4934, min_samples_split=50, n_estimators=1927)
# fit the model
model.fit(X_train, y_train)
# get importance
importance = model.feature_importances_
sorted_idx = model.feature_importances_.argsort()
# summarize feature importance
for i,v in enumerate(importance[sorted_idx]):
	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
plt.barh(datasetF.columns[sorted_idx], model.feature_importances_[sorted_idx])
plt.rcParams['figure.figsize'] = [20, 20]
plt.rcParams.update({'font.size': 20})
plt.tight_layout()
plt.savefig("FeatureImportance.png", dpi = 1200)
plt.show()
```
